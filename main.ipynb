```python

import os
import PIL
from PIL import Image
from skimage import io
import tensorflow as tf

import pathlib
```

```python
import os
import numpy as np

import matplotlib.pyplot as plt
import cv2
```

```python
path1 = "Image sets/Normal"
images1 = os.listdir(path1)
images1
```

```python
imgs = []
for image in images1:
    img = Image.open(os.path.join(path1,image))
    img = np.asarray(img)
    imgs.append(img)
imgs = np.array(imgs)
```

```python
num_rows = 10
num_cols = 15
num_images = num_rows*num_cols
rand_indices = np.random.choice(len(imgs), num_images, replace=False)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 20))
for i, ax in zip(rand_indices, axes.flat):
    ax.imshow(imgs[i])
    ax.axis('off')
plt.show()
```
